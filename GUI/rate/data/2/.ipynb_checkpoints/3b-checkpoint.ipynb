### Bistability

```python
aln = ALNModel()
N = aln.params.N

data.set_parameters(aln)

state_vars = aln.state_vars
init_vars = aln.init_vars

##############################################################
def setinit(init_vars_, model):
    state_vars = model.state_vars
    init_vars = model.init_vars
    for iv in range(len(init_vars)):
        for sv in range(len(state_vars)):
            if state_vars[sv] in init_vars[iv]:
                #print("set init vars ", )
                if model.params[init_vars[iv]].ndim == 2:
                    model.params[init_vars[iv]][0,:] = init_vars_[sv]
                else:
                    model.params[init_vars[iv]][0] = init_vars_[sv]
                    
##############################################################               
def setmaxmincontrol(max_c_c, min_c_c, max_c_r, min_c_r):
    import numpy as np
    
    max_cntrl = np.zeros(( 6 ))
    min_cntrl = np.zeros(( 6 ))
    
    max_cntrl[0] = max_c_c
    min_cntrl[0] = min_c_c
    max_cntrl[1] = max_c_c
    min_cntrl[1] = min_c_c
    max_cntrl[2] = max_c_r
    min_cntrl[2] = min_c_r
    max_cntrl[3] = max_c_r
    min_cntrl[3] = min_c_r
    max_cntrl[4] = max_c_r
    min_cntrl[4] = min_c_r
    max_cntrl[5] = max_c_r
    min_cntrl[5] = min_c_r
            
    return max_cntrl, min_cntrl
```

```python
d_array = np.arange(20., 300, 0.1)
t0_array = d_array - 20.
t_pen = np.zeros(( t0_array.shape ))
for i in range(len(t_pen)):
    t_pen[i] = t0_array[i] / d_array[i]

bestControl_ = np.array( [[None] * len(t_pen)] * len(exc) )
bestState_ = np.array( [[None] * len(t_pen)] * len(exc) )
cost_ = np.array( [[None] * len(t_pen)] * len(exc) )
costnode_ = np.array( [[None] * len(t_pen)] * len(exc) )
weights_ = np.array( [[None] * len(t_pen)] * len(exc) )
convergence_ = np.array( [[None] * len(t_pen)] * len(exc) )

initVars = [None] * len(exc)
target = [None] * len(exc)
cost_uncontrolled = [None] * len(exc)

print(bestControl_.shape)
```

```python
file_ = '3b.pickle'
case = '1'
```

```python
dur_pre = 10
dur_post = 10

n_pre = int(np.around(dur_pre/aln.params.dt + 1.,1))
n_post = int(np.around(dur_post/aln.params.dt + 1.,1))

tol = 1e-32
start_step = 10.
c_scheme = np.zeros(( 1,1 ))
c_scheme[0,0] = 1.
u_mat = np.identity(1)
u_scheme = np.array([[1.]])

c_var = [0,1]
p_var = [0]

if case in ['1', '2']:    # low to high
    max_I = [3., -3.]
else:
    max_I = [-3., 3.]
    
if case in ['1', '3']:    # sparsity
    factor_ws = 1.
    factor_we = 0.
else:
    factor_ws = 0.
    factor_we = 1.
    
maxC = [5., -5., 0.18, 0.]
max_cntrl, min_cntrl = setmaxmincontrol(maxC[0], maxC[1], maxC[2], maxC[3])
```

```python
if os.path.isfile(file_) :
    print("file found")
    
    with open(file_,'rb') as f:
        load_array = pickle.load(f)

    bestControl_ = load_array[0]
    bestState_ = load_array[1]
    cost_ = load_array[2]
    costnode_ = load_array[3]
    weights_ = load_array[4]
```

```python
# get initial parameters and target states

#i_stepsize = 20
#limit = 150
#start_index = 0
#i_range = range(start_index, limit,i_stepsize)
i_range = [20]

data.set_parameters(aln)

for i in i_range:
    print("------- ", i, exc[i], inh[i])
    aln.params.ext_exc_current = exc[i] * 5.
    aln.params.ext_inh_current = inh[i] * 5.
    
    aln.params.duration = 3000.
    
    control0 = aln.getZeroControl()
    control0 = functions.step_control(aln, maxI_ = max_I[0])

    aln.run(control=control0)
    
    target_rates = np.zeros((2))
    target_rates[0] = aln.rates_exc[0,-1] 
    target_rates[1] = aln.rates_inh[0,-1]

    control0 = functions.step_control(aln, maxI_ = max_I[1])
    aln.run(control=control0)

    init_state_vars = np.zeros(( len(state_vars) ))
    for j in range(len(state_vars)):
        if aln.state[state_vars[j]].size == 1:
            init_state_vars[j] = aln.state[state_vars[j]][0]
        else:
            init_state_vars[j] = aln.state[state_vars[j]][0,-1]

    initVars[i] = init_state_vars
    target[i] = [target_rates[0], target_rates[1]]
```

```python
# get uncontrolled cost

data.set_parameters(aln)

for i in i_range:
    print("------- ", i, exc[i], inh[i])
    aln.params.ext_exc_current = exc[i] * 5.
    aln.params.ext_inh_current = inh[i] * 5.
    
    dur = d_array[0]
    aln.params.duration = dur
    
    target_ = aln.getZeroTarget()
    target_[:,0,:] = target[i][0]
    target_[:,1,:] = target[i][1]
        
    cost.setParams(1.0, 0.0, 0.0)

    setinit(initVars[i], aln)
    control0 = aln.getZeroControl()

    # "HS", "FR", "PR", "HZ"
    cgv = None
    max_it = 0

    bestControl_init_, bestState_init_, cost_init_, runtime_init_, grad_init_, phi_init_, costnode_init_ = aln.A1(
        control0, target_, c_scheme, u_mat, u_scheme, max_iteration_ = max_it, tolerance_ = tol,
        startStep_ = start_step, max_control_ = max_cntrl, min_control_ = min_cntrl, t_sim_ = dur,
        t_sim_pre_ = dur_pre, t_sim_post_ = dur_post, CGVar = cgv, control_variables_ = c_var,
        prec_variables_ = p_var, transition_time_ = 0.)
    
    cost_uncontrolled[i] = cost_init_[0]
```

```python
cgv = None
j = 0
k = 0

while k < 10:
    print('-------------------------------------------- k = ', k)

    for i in i_range:
        print("------- ", i, exc[i], inh[i])
        aln.params.ext_exc_current = exc[i] * 5.
        aln.params.ext_inh_current = inh[i] * 5.

        if convergence_[i][j] == True:
            print('converged')
            continue

        dur = d_array[0]
        aln.params.duration = dur
        
        setinit(initVars[i], aln)

        target_ = aln.getZeroTarget()
        target_[:,0,:] = target[i][0]
        target_[:,1,:] = target[i][1]

        max_it = int(100)

        if type(bestControl_[i][j]) == type(None) or np.amax(np.abs(bestControl_[i][j])) < 1e-6:
            print('no solution found')
            control0 = aln.getZeroControl()
            #control0[0,0,:50] = 5.
            cost.setParams(1.0, 0., 1.)
            max_it = int(1)
            weights_[i][j] = [1., 0., 1.]
        else:
            control0 = bestControl_[i][j][:,:,n_pre-1:-n_post+1]
            l = 1
            while cost_[i][j][-l] == 0.:
                l += 1

            weight_ = ( cost_uncontrolled[i] - costnode_[i][j][0][0][0] ) / sum( costnode_[i][j][2][0][:] )
                
            print("weight = ", weight_)
            cost.setParams(1.0, weight_ * factor_we, weight_ * factor_ws)
            weights_[i][j] = cost.getParams()

        bestControl_[i][j], bestState_[i][j], cost_[i][j], runtime_, grad_, phi_, costnode_[i][j] = aln.A1(
            control0, target_, c_scheme, u_mat, u_scheme, max_iteration_ = max_it, tolerance_ = tol,
            startStep_ = start_step, max_control_ = max_cntrl, min_control_ = min_cntrl, t_sim_ = dur,
            t_sim_pre_ = dur_pre, t_sim_post_ = dur_post, CGVar = cgv, control_variables_ = c_var,
            prec_variables_ = p_var, transition_time_ = t_pen[j])

        #if cost_[i][j][2] == 0.:
        #    print('converged')
        #    convergence_[i][j] = True

        print('uncontrolled cost : ', cost_uncontrolled[i])
        print('cost splitting : ', costnode_[i][j][0][0][0], costnode_[i][j][2][0][:2])

        time_ = np.arange(-dur_pre,dur+dur_post+aln.params.dt,aln.params.dt)
        dur += dur_pre + dur_post
        target_ = aln.getZeroTarget()
        target_[:,0,:] = target[i][0]
        target_[:,1,:] = target[i][1]

        if True:
            fig, ax = plt.subplots(2,1, constrained_layout=True, figsize = (4,4) )
            ax[1].plot(time_, bestControl_[i][j][0,0,:], color = 'red')
            ax[0].plot(time_, bestState_[i][j][0,0,:], color= 'red')
            ax[1].plot(time_, bestControl_[i][j][0,1,:], color = 'blue')
            ax[0].plot(time_, bestState_[i][j][0,1,:], color= 'blue')
            ax[0].plot(time_, target_[0,0,:], linestyle = '--', color='red')
            ax[0].plot(time_, target_[0,1,:], linestyle = '--', color='blue')
            ax[0].set_xlim(-dur_pre,dur-dur_post)
            ax[1].set_xlim(-dur_pre,dur-dur_post)
            ax[0].axvspan(t0_array[0], dur, alpha=0.2, color='gray')
            plt.show()

        with open(file_,'wb') as f:
            pickle.dump([bestControl_, bestState_, cost_, costnode_, weights_], f)
    
    k += 1
```

```python
for i in range(bestControl_.shape[0]):
    for j in range(1,bestControl_.shape[1]):
        bestControl_[i][j] = None
        convergence_[i][j] = None
```

```python
cost.setParams(1.0, factor_we, factor_ws)

j_start = 1

for i in i_range:
    print("------- ", i, exc[i], inh[i])
    aln.params.ext_exc_current = exc[i] * 5.
    aln.params.ext_inh_current = inh[i] * 5.

    for j in range(j_start,j_start+100):
        print(i,j, '-----------------')
            
        k = 0
        while convergence_[i][j] != True:
            print('---------------------', k)

            dur = round(d_array[j],1)
            print('simulation duration : ', dur)
            aln.params.duration = dur
            max_it = int(100)  
            
            setinit(initVars[i], aln)
            target_ = aln.getZeroTarget()
            target_[:,0,:] = target[i][0]
            target_[:,1,:] = target[i][1]

            if type(bestControl_[i][j]) == type(None):
                print('no solution computed yet')
                control0 = aln.getZeroControl()
                control1 = aln.getZeroControl()
                if j%2 == 0:
                    control0[:,:,1:] = bestControl_[i][j-1][:,:,n_pre-1:-n_post+1]
                else:
                    control0[:,:,:-1] = bestControl_[i][j-1][:,:,n_pre-1:-n_post+1]
                    
                time_ = np.arange(0.,round(dur+aln.params.dt,1),aln.params.dt)
                basis, coeffs = smoothfit.fit1d(time_[1:], control0[0,0,1:],
                                aln.params.dt, round(dur+aln.params.dt,1), int(len(time_)-2), degree=1, lmbda=1.)
                control1[0,0,1:] = coeffs[basis.nodal_dofs[0]]
                basis, coeffs = smoothfit.fit1d(time_[1:], control0[0,1,1:],
                                aln.params.dt, round(dur+aln.params.dt,1), int(len(time_)-2), degree=1, lmbda=1.)
                control1[0,1,1:] = coeffs[basis.nodal_dofs[0]]                            
                                        
                l = 1
                while cost_[i][j-1][-l] == 0.:
                    l += 1
                    
                weight_ = ( cost_uncontrolled[i] - costnode_[i][j-1][0][0][0] ) / sum( costnode_[i][j-1][2][0][:] )
            else:
                control1 = bestControl_[i][j][:,:,n_pre-1:-n_post+1]
                l = 1
                while cost_[i][j][-l] == 0.:
                    l += 1
                
                weight_ = ( cost_uncontrolled[i] - costnode_[i][j][0][0][0] ) / sum( costnode_[i][j][2][0][:] )
                
            print("weight = ", weight_)
            cost.setParams(1.0, weight_ * factor_we, weight_ * factor_ws)
            weights_[i][j] = cost.getParams()

            bestControl_[i][j], bestState_[i][j], cost_[i][j], runtime_, grad_, phi_, costnode_[i][j] = aln.A1(
                control1, target_, c_scheme, u_mat, u_scheme, max_iteration_ = max_it, tolerance_ = tol,
                startStep_ = start_step, max_control_ = max_cntrl, min_control_ = min_cntrl, t_sim_ = dur,
                t_sim_pre_ = dur_pre, t_sim_post_ = dur_post, CGVar = cgv, control_variables_ = c_var,
                prec_variables_ = p_var, transition_time_ = t_pen[j])

            if cost_[i][j][2] == 0.:
                print('converged')
                convergence_[i][j] = True

            print('uncontrolled cost : ', cost_uncontrolled[i])
            print('cost splitting : ', costnode_[i][j][0][0][0], costnode_[i][j][2][0][:2])
                
            time_ = np.arange(-dur_pre,round(dur+dur_post+aln.params.dt,1),aln.params.dt)
            dur += dur_pre + dur_post
            target_ = aln.getZeroTarget()
            target_[:,0,:] = target[i][0]
            target_[:,1,:] = target[i][1]

            if False:
                fig, ax = plt.subplots(2,1, constrained_layout=True, figsize = (4,4) )
                ax[1].plot(time_, bestControl_[i][j][0,0,:], color = 'red')
                ax[0].plot(time_, bestState_[i][j][0,0,:], color= 'red')
                ax[1].plot(time_, bestControl_[i][j][0,1,:], color = 'blue')
                ax[0].plot(time_, bestState_[i][j][0,1,:], color= 'blue')
                ax[0].plot(time_, target_[0,0,:], linestyle = '--', color='red')
                ax[0].plot(time_, target_[0,1,:], linestyle = '--', color='blue')
                ax[0].set_xlim(-dur_pre,dur-dur_post)
                ax[1].set_xlim(-dur_pre,dur-dur_post)
                ax[0].axvspan(t0_array[j], t0_array[j]+20, alpha=0.2, color='gray')
                plt.show()

            with open(file_,'wb') as f:
                pickle.dump([bestControl_, bestState_, cost_, costnode_, weights_], f)
                
            k += 1
```
