```python
### w = 1

i = 20

file_I = '.' + os.sep + 'data_2' + os.sep + '1' + os.sep + '1_I.pickle'
final_file_tradeoff = '.' + os.sep + 'data_final_2' + os.sep + 'control_1_tradeoff_' + str(i) + '_W1.pickle'
file_up = '.' + os.sep + 'data_2' + os.sep + '1' + os.sep + '1a_W1.pickle'

if os.path.isfile(file_I) :
    print("file found")

    with open(file_I,'rb') as f:
        load_array = pickle.load(f)

    bestControl_0 = load_array[0][i]
    bestState_0 = load_array[1][i]
    cost_0 = load_array[2][i]
    costnode_0 = load_array[3][i]
    weights_0 = load_array[4][i]

if os.path.isfile(file_up) :
    print("file found")

    with open(file_up,'rb') as f:
        load_array = pickle.load(f)

    bestControl_1 = load_array[0][i]
    bestState_1 = load_array[1][i]
    cost_1 = load_array[2][i]
    costnode_1 = load_array[3][i]
    weights_1 = load_array[4][i]

bestControl = [None] * (len(bestControl_0) + 1)
bestState = [None] * (len(bestControl_0) + 1)
cost = [None] * (len(bestControl_0) + 1)
costnode = [None] * (len(bestControl_0) + 1)
weights = [None] * (len(bestControl_0) + 1)

bestControl[len(bestControl_0)] = bestControl_1[0]
bestState[len(bestControl_0)] = bestState_1[0]
cost[len(bestControl_0)] = cost_1[0]
costnode[len(bestControl_0)] = costnode_1[0]
weights[len(bestControl_0)] = weights_1[0]

for j in range(len(bestControl_0)):
    
    j_ = len(bestControl_0)
    print(j, j_ - j)
    
    if type(bestControl_0[j]) == type(None):
        if j < 2001:
            continue
        if type(bestControl_1[j_-j]) == type(None):
            continue     
    
    if j < 2001:
        bestControl[j] = bestControl_0[j]
        bestState[j] = bestState_0[j]
        cost[j] = cost_0[j]
        costnode[j] = costnode_0[j]
        weights[j] = weights_0[j]
        continue
        
    if type(bestControl_1[j_-j]) == type(None):
        bestControl[j] = bestControl_0[j]
        bestState[j] = bestState_0[j]
        cost[j] = cost_0[j]
        costnode[j] = costnode_0[j]
        weights[j] = weights_0[j]
        continue
        
    if type(bestControl_0[j]) == type(None):
        bestControl[j] = bestControl_1[j_-j]
        bestState[j] = bestState_1[j_-j]
        cost[j] = cost_1[j_-j]
        costnode[j] = costnode_1[j_-j]
        weights[j] = weights_1[j_-j]
        continue
    
    cost0 = sum(costnode_0[j][0][0][:]) + sum(costnode_0[j][2][0][:])
    cost1 = sum(costnode_1[j_-j][0][0][:]) + sum(costnode_1[j_-j][2][0][:])

    print(cost0, cost1)

    if np.amin([cost0, cost1]) == cost0:
        print(j, ' downwards ')
        bestControl[j] = bestControl_0[j]
        bestState[j] = bestState_0[j]
        cost[j] = cost_0[j]
        costnode[j] = costnode_0[j]
        weights[j] = weights_0[j]

    elif np.amin([cost0, cost1]) == cost1:
        print(j, ' upwards')
        bestControl[j] = bestControl_1[j_-j]
        bestState[j] = bestState_1[j_-j]
        cost[j] = cost_1[j_-j]
        costnode[j] = costnode_1[j_-j]
        weights[j] = weights_1[j_-j]

with open(final_file_tradeoff,'wb') as f:
    pickle.dump([bestControl, bestState, cost, costnode, weights], f)
```
