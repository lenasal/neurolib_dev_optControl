```python
def F(x, z):
    
    F = np.zeros(( x.shape[0] ))
    
    F[0] = ( - x[0] + mu_ext_e + J_ee * x[2] + J_ei * x[3] ) / z[4]
    ree = rxy(cee, Ke, tau_se, z[0], J_ee)
    F[2] =  - x[2] / tau_se + (1. - x[2]) * ree / tau_se
    pee = pxy(cee, Ke, tau_se, z[0], J_ee)
    F[6] =  ( (1. - x[2])**2 * pee + ( pee - 2. * tau_se * ( ree + 1. )) * max(0., x[6]) ) / tau_se**2
    
    F[1] = ( - x[1] + mu_ext_i + J_ii * x[5] + J_ie * x[4] ) / z[5]
    rii = rxy(cii, Ki, tau_si, z[1], J_ii)
    F[5] =  - x[5] / tau_si + (1. - x[5]) * rii / tau_si
    pii = pxy(cii, Ki, tau_si, z[1], J_ii)
    F[9] =  ( (1. - x[5])**2 * pii + ( pii - 2. * tau_si * ( rii + 1. )) * max(0., x[9]) ) / tau_si**2
    
    rei = rxy(cei, Ki, tau_si, z[1], J_ei)
    F[3] = - x[3] / tau_si + (1. - x[3]) * rei / tau_si
    pei = pxy(cei, Ki, tau_si, z[1], J_ei)
    F[7] =  ( (1. - x[3])**2 * pei + ( pei - 2. * tau_si * ( rei + 1. )) * max(0., x[7]) ) / tau_si**2
    
    rie = rxy(cie, Ke, tau_se, z[0], J_ie)
    F[4] = - x[4] / tau_se + (1. - x[4]) * rie / tau_se
    pie = pxy(cie, Ke, tau_se, z[0], J_ie)
    F[8] =  ( (1. - x[4])**2 * pie + ( pie - 2. * tau_se * ( rie + 1. )) * max(0., x[8]) ) / tau_se**2
    
    return F

def G(x, z):
    G = np.zeros(( z.shape[0] ))
    G[0] = - interpolate(r_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma) * 1e3
    ree = rxy(cee, Ke, tau_se, z[0], J_ee)
    rei = rxy(cei, Ki, tau_si, z[1], J_ei)
    G[2] = - np.sqrt( 2. * J_ee**2 * x[6] * tau_se * taum / ( (1. + ree ) * taum + tau_se )
                    + 2. * J_ei**2 * x[7] * tau_si * taum / ( (1. + rei ) * taum + tau_si )
                    + sigma_ext_e**2 )
    G[4] = - interpolate(tau_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma)
    G[1] = - interpolate(r_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma) * 1e3
    rii = rxy(cii, Ki, tau_si, z[1], J_ii)
    rie = rxy(cie, Ke, tau_se, z[0], J_ie)
    G[3] = - np.sqrt( 2. * J_ii**2 * x[9] * tau_si * taum / ( (1. + rii ) * taum + tau_si )
                    + 2. * J_ie**2 * x[8] * tau_se * taum / ( (1. + rie ) * taum + tau_se )
                    + sigma_ext_i**2 )
    G[5] = - interpolate(tau_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma)
    return G
```

```python
def interpolate(table_, mu_, sigma_, mu_range_, d_mu_, sigma_range_, d_sigma_):
    xid1, yid1, dxid, dyid = timeIntegration.fast_interp2_opt(sigma_range_, d_sigma_, sigma_, mu_range_, d_mu_, mu_)
    xid1, yid1 = int(xid1), int(yid1)
    return timeIntegration.interpolate_values(table_, xid1, yid1, dxid, dyid)
    
def rate_e(x, z):
    return interpolate(r_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma) * 1e3

def rate_i(x, z):
    return interpolate(r_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma) * 1e3

def tau_e(x, z):
    return interpolate(tau_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma)

def tau_i(x, z):
    return interpolate(tau_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma)

def rxy(cxy, Ky, tau_sy, ry, J_xy):
    return cxy * Ky * tau_sy * ry * 1e-3 / np.abs(J_xy)

def pxy(cxy, Ky, tau_sy, ry, J_xy):
    return cxy**2 * Ky * tau_sy**2 * ry * 1e-3 / J_xy**2
```

```python
def F_dx(x, z, dx):
    F_dx = np.zeros(( x.shape[0], x.shape[0] ))
    for i in range(x.shape[0]):
        for j in range(F_dx.shape[1]):
            x_dx = x.copy()
            x_dx[j] = x[j] + dx
            F_1 = F(x_dx, z)
            F_0 = F(x, z)
            F_dx[i,j] = ( F_1[i] - F_0[i] ) / (dx)
    return F_dx

def F_dz(x, z, dz):
    F_dz = np.zeros(( x.shape[0], z.shape[0] ))
    for i in range(F_dz.shape[0]):
        for j in range(F_dz.shape[1]):
            z_dz = z.copy()
            z_dz[j] = z[j] + dz
            F_1 = F(x, z_dz)
            F_0 = F(x, z)
            F_dz[i,j] = ( F_1[i] - F_0[i] ) / (dz)
    return F_dz

def G_dx(x, z, dx):
    G_dx = np.zeros(( z.shape[0], x.shape[0] ))
    for i in range(G_dx.shape[0]):
        for j in range(G_dx.shape[1]):
            x_dx = x.copy()
            x_dx[j] = x[j] + dx
            G_1 = G(x_dx, z)
            G_0 = G(x, z)
            G_dx[i,j] = ( G_1[i] - G_0[i] ) / (dx)
    return G_dx

def G_dz(x, z, dz):
    G_dz = np.zeros(( z.shape[0], z.shape[0] ))
    for i in range(G_dz.shape[0]):
        for j in range(G_dz.shape[1]):
            z_dz = z.copy()
            z_dz[j] = z[j] + dz
            G_1 = G(x, z_dz)
            G_0 = G(x, z)
            G_dz[i,j] = ( G_1[i] - G_0[i] ) / (dz)
        G_dz[i,i] += 1.
    return G_dz
```

```python
def H(x, z, dx, dz):
    H0 = G_dz(x, z, dz)
    H1 = np.linalg.inv(H0)
    #print(H1)
    H2 = G_dx(x, z, dx)
    H3 = F(x, z)
    H12 = np.tensordot(H1, H2, axes=1)
    H123 = np.tensordot(H12, H3, axes=1)
    
    #print(H1)
    #print('---')
    #print(H2)
    #print('---')
    #print(H3)
    #print('---')
    
    return - H123

def H_dx(x, z, dx, dz):
    H_dx = np.zeros((dim_z, dim_x))
    for i in range(H_dx.shape[0]):
        for j in range(H_dx.shape[1]):
            x_dx = x.copy()
            x_dx[j] = x[j] + dx
            H_1 = H(x_dx, z, dx, dz)
            H_0 = H(x, z, dx, dz)
            H_dx[i,j] = ( H_1[i] - H_0[i] ) / (dx)
    return H_dx

def H_dz(x, z, dx, dz):
    H_dz = np.zeros((dim_z, dim_z))
    for i in range(H_dz.shape[0]):
        for j in range(H_dz.shape[1]):
            z_dz = z.copy()
            z_dz[j] = z[j] + dz
            H_1 = H(x, z_dz, dx, dz)
            H_0 = H(x, z, dx, dz)
            H_dz[i,j] = ( H_1[i] - H_0[i] ) / (dz)
    return H_dz

def jacobian(x, z, dx):

    F1 = F_dx(x,z, dx)
    F2 = F_dz(x, z, dx)
    #H0 = H(x, z, dx, dx)
    H1 = H_dx(x, z, dx, dx)
    H2 = H_dz(x, z, dx, dx)

    jac = np.zeros((dim_x+dim_z, dim_x+dim_z))
    jac[:dim_x,:dim_x] = F1
    jac[dim_x:, :dim_x] = H1
    jac[:dim_x, dim_x:] = F2
    jac[dim_x:, dim_x:] = H2
        
    return jac

def B_(z_):
    B = np.zeros(( dim_v, dim_v ))
    tauE = z_[4]
    tauI = z_[5]
    B[0,0] = 1./tauE
    B[1,1] = 1./tauI
    return B
```

```python
def get_x_z_v(state_, t_):
    x_ = np.zeros(( dim_x ))
    z_ = np.zeros(( dim_z ))
    v_ = np.zeros(( dim_v ))
    
    x_[0] = state_[0,2,t_]
    x_[1] = state_[0,3,t_]
    x_[2] = state_[0,5,t_]
    x_[3] = state_[0,6,t_]
    x_[4] = state_[0,7,t_]
    x_[5] = state_[0,8,t_]
    x_[6] = state_[0,9,t_]
    x_[7] = state_[0,10,t_]
    x_[8] = state_[0,11,t_]
    x_[9] = state_[0,12,t_]
    
    z_[0] = state_[0,0,t_]
    z_[1] = state_[0,1,t_]
    z_[2] = state_[0,15,t_]
    z_[3] = state_[0,16,t_]
    z_[4] = state_[0,18,t_]
    z_[5] = state_[0,19,t_]
    
    v_[:dim_x] = x_[:]
    v_[dim_x:] = z_[:]
    
    return x_, z_, v_
```

```python
def lin_forward(E_, A_, A_E_, A_I_,  B_, C_, x0, u_, T_, dt_):
    x_ = np.zeros(( A_.shape[0], int(T_/dt + x0.shape[2] ) ))
    #print(x_.shape)
    shift_e = int(aln.params.de/dt_)
    shift_i = int(aln.params.di/dt_)
    x_[:,:max(shift_e,shift_i)+1] = x0[:,:]
    
    #print(x_[:,:])
    #print('----')
    
    for t_ in range(max(shift_e,shift_i)+1, x_.shape[1]):
        
        #print(t_)
        
        x_[:,t_] = ( np.matmul(A_, x_[:,t_-1]) + np.matmul(A_E_, x_[:,t_-1-shift_e])
                    + np.matmul(A_I_, x_[:,t_-1-shift_i]) + np.matmul(B_, u_[:,t_-max(shift_e,shift_i)-1]) + C_[:] ) *dt + x_[:,t_-1]
        
        for j in range(E_.shape[0]):
            #print(E_[j], x_[j,t_])
            x_[j,t_] *= E_[j]
            #print(E_[j], x_[j,t_])
            
        #j = 15
        #print(j, np.matmul(A_[j,:], x_[:,t_]), np.matmul(A_E_[j,:], x_[:,t_-shift_e]), C_[j])
        #print(A_[j,:])
        #print(x_[j,t_])
                
        if shift_e > 0. and shift_i > 0.:
            res = ( np.matmul(A_, x_[:,t_]) + np.matmul(A_E_, x_[:,t_-shift_e])
                        + np.matmul(A_I_, x_[:,t_-shift_i]) + np.matmul(B_, u_[:,t_-max(shift_e,shift_i)]) + C_[:] )
        
        elif shift_e == 0. and shift_i > 0.:
            res = ( np.matmul(A_, x_[:,t_]) + np.matmul(A_E_, x_[:,t_-shift_e-1])
                    + np.matmul(A_I_, x_[:,t_-shift_i]) + np.matmul(B_, u_[:,t_-max(shift_e,shift_i)]) + C_[:] )
        elif shift_i == 0. and shift_e > 0.:
            res = ( np.matmul(A_, x_[:,t_]) + np.matmul(A_E_, x_[:,t_-shift_e])
                        + np.matmul(A_I_, x_[:,t_-shift_i-1]) + np.matmul(B_, u_[:,t_-max(shift_e,shift_i)]) + C_[:] )
        else:
            res = ( np.matmul(A_, x_[:,t_]) + np.matmul(A_E_, x_[:,t_-shift_e-1])
                        + np.matmul(A_I_, x_[:,t_-shift_i-1]) + np.matmul(B_, u_[:,t_-max(shift_e,shift_i)]) + C_[:] )
        
        for j in [15,16]:
            x_[j,t_] = - res[j] / A_[j,j]
            #x_[15,t_] = 1.5757341
            #x_[16,t_] = 1.53542308
            
        res = ( np.matmul(A_, x_[:,t_]) + np.matmul(A_E_, x_[:,t_-shift_e])
                    + np.matmul(A_I_, x_[:,t_-shift_i]) + np.matmul(B_, u_[:,t_-max(shift_e,shift_i)]) + C_[:] )
        
        for j in [0,1,18,19]:
            x_[j,t_] = - res[j] / A_[j,j]
            
    return x_
```

```python
def x_u_sol(B_, C_, V_, R_, T_, jac_u_, dt_):
    x_sol = np.zeros(( 2*dim_v, int(T_/dt + 1) )).astype(complex)
    u_sol = np.zeros(( dim_v , int(T_/dt + 1) )).astype(complex)
        
    for t_ in range(x_sol.shape[1]):
        for j in range(2*dim_v):
            x_sol[:,t_] += B_[j] * V_[:,j] * np.exp(R_[j] * t_ * dt)
        x_sol[:,t_] += C_[:]
        
        u_sol[:,t_] = - ( 1. / w2 ) * np.matmul( np.transpose(jac_u_), x_sol[dim_v:, t_] )
    
    return x_sol, u_sol
```

```python
def crosscheck(x_crosscheck, jac_, jac_u_, u, dt):
    
    for t_ in range(1, x_crosscheck.shape[1]):
        for j in range(dim_v):
            dxj = 0.
            for k in range(dim_v):
                dxj += ( jac_[j,k] * x_crosscheck[k,t_-1] )
                dxj += ( jac_u_[j,k] * u[k,t_-1] )

            x_crosscheck[j,t_] += dxj * dt + x_crosscheck[j,t_-1]
            
    return x_crosscheck
```
