```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import pickle
import random

path = os.getcwd().split(os.sep + 'GUI')[0]
if path not in sys.path:
    sys.path.append(path)

from neurolib.models.aln import ALNModel
import neurolib.dashboard.data as data
from neurolib.models.aln import timeIntegration

%load_ext autoreload
%autoreload 2
```

```python
dim_x = 10
dim_z = 6

aln = ALNModel()

r_table = aln.params.precalc_r
tau_table = aln.params.precalc_tau_mu
mu_range = aln.params.Irange
d_mu = aln.params.dI
sigma_range = aln.params.sigmarange
d_sigma = aln.params.ds

J_ee = aln.params.Jee_max
J_ei = aln.params.Jei_max
J_ie = aln.params.Jie_max
J_ii = aln.params.Jii_max

cee = aln.params.cee
cei = aln.params.cei
cie = aln.params.cie
cii = aln.params.cii

sigma_ext_e = aln.params.sigmae_ext
sigma_ext_i = aln.params.sigmai_ext
mu_ext_e = aln.params.ext_inh_current
mu_ext_i = aln.params.ext_exc_current

Ke = aln.params.Ke
Ki = aln.params.Ki
tau_se = aln.params.tau_se
tau_si = aln.params.tau_si
taum = aln.params.C / aln.params.gL

mu_ext_e = 0.45 * 5.
mu_ext_i = 0.475 * 5.
```

```python
def setinithigh(x_timeseries, z_timeseries):
    x_timeseries[0,0] = 2.
    x_timeseries[1,0] = 0.
    x_timeseries[2,0] = 0.5
    x_timeseries[3,0] = 0.
    x_timeseries[4,0] = 0.
    x_timeseries[5,0] = 0.
    x_timeseries[6,0] = 0.2
    x_timeseries[7,0] = 0.4
    x_timeseries[8,0] = 0.6
    x_timeseries[9,0] = 0.8

    z_timeseries[0,0] = 0.
    z_timeseries[1,0] = 0.
    z_timeseries[2,0] = 0.
    z_timeseries[3,0] = 0.
    z_timeseries[4,0] = 5.
    z_timeseries[5,0] = 10.
    
def setinitlow(x_timeseries, z_timeseries):
    x_timeseries[0,0] = -0.3
    x_timeseries[1,0] = 1.
    x_timeseries[2,0] = 0.1
    x_timeseries[3,0] = 1.
    x_timeseries[4,0] = 0.1
    x_timeseries[5,0] = 1.
    x_timeseries[6,0] = 0.
    x_timeseries[7,0] = 0.
    x_timeseries[8,0] = 0.
    x_timeseries[9,0] = 0.

    z_timeseries[0,0] = 5.
    z_timeseries[1,0] = 10.
    z_timeseries[2,0] = 0.1
    z_timeseries[3,0] = 0.1
    z_timeseries[4,0] = 20.
    z_timeseries[5,0] = 1.
    
def setinitrandom(x_timeseries, z_timeseries):
    x_timeseries[0,0] = random.uniform(-.5, 4.)
    x_timeseries[1,0] = random.uniform(-.5, 4.)
    x_timeseries[2,0] = random.uniform(0.1, 1.)
    x_timeseries[3,0] = random.uniform(0.1, 1.)
    x_timeseries[4,0] = random.uniform(0.1, 1.)
    x_timeseries[5,0] = random.uniform(0.1, 1.)
    x_timeseries[6,0] = random.uniform(0.1, 1.)
    x_timeseries[7,0] = random.uniform(0.1, 1.)
    x_timeseries[8,0] = random.uniform(0.1, 1.)
    x_timeseries[9,0] = random.uniform(0.1, 1.)

    z_timeseries[0,0] = random.uniform(1., 70.)
    z_timeseries[1,0] = random.uniform(1., 100.)
    z_timeseries[2,0] = random.uniform(0.1, 1.)
    z_timeseries[3,0] = random.uniform(0.1, 1.)
    z_timeseries[4,0] = random.uniform(2., 10.)
    z_timeseries[5,0] = random.uniform(2., 10.)

def ufp():
    
    x_ufp = np.array([4.60978687e-01, 2.04701135e+00, 4.89328731e-01, 9.01902320e-01,
       4.72445380e-01, 9.48719719e-01, 8.12620177e-03, 7.05792669e-04, 7.80202698e-03, 4.44632506e-04])

    z_ufp = np.array([4.87249427, 60.70883737, 1.53316218, 1.53502226, 9.23210462, 0.81208368])
    
    return x_ufp, z_ufp
```

```python
def F(x, z):
    
    F = np.zeros(( x.shape[0] ))
    
    F[0] = ( - x[0] + mu_ext_e + J_ee * x[2] + J_ei * x[3] ) / z[4]
    ree = rxy(cee, Ke, tau_se, z[0], J_ee)
    F[2] =  - x[2] / tau_se + (1. - x[2]) * ree / tau_se
    pee = pxy(cee, Ke, tau_se, z[0], J_ee)
    F[6] =  ( (1. - x[2])**2 * pee + ( pee - 2. * tau_se * ( ree + 1. )) * max(0., x[6]) ) / tau_se**2
    
    F[1] = ( - x[1] + mu_ext_i + J_ii * x[5] + J_ie * x[4] ) / z[5]
    rii = rxy(cii, Ki, tau_si, z[1], J_ii)
    F[5] =  - x[5] / tau_si + (1. - x[5]) * rii / tau_si
    pii = pxy(cii, Ki, tau_si, z[1], J_ii)
    F[9] =  ( (1. - x[5])**2 * pii + ( pii - 2. * tau_si * ( rii + 1. )) * max(0., x[9]) ) / tau_si**2
    
    rei = rxy(cei, Ki, tau_si, z[1], J_ei)
    F[3] = - x[3] / tau_si + (1. - x[3]) * rei / tau_si
    pei = pxy(cei, Ki, tau_si, z[1], J_ei)
    F[7] =  ( (1. - x[3])**2 * pei + ( pei - 2. * tau_si * ( rei + 1. )) * max(0., x[7]) ) / tau_si**2
    
    rie = rxy(cie, Ke, tau_se, z[0], J_ie)
    F[4] = - x[4] / tau_se + (1. - x[4]) * rie / tau_se
    pie = pxy(cie, Ke, tau_se, z[0], J_ie)
    F[8] =  ( (1. - x[4])**2 * pie + ( pie - 2. * tau_se * ( rie + 1. )) * max(0., x[8]) ) / tau_se**2
    
    return F

def G(x, z):
    G = np.zeros(( z.shape[0] ))
    G[0] = - interpolate(r_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma) * 1e3
    ree = rxy(cee, Ke, tau_se, z[0], J_ee)
    rei = rxy(cei, Ki, tau_si, z[1], J_ei)
    G[2] = - np.sqrt( 2. * J_ee**2 * x[6] * tau_se * taum / ( (1. + ree ) * taum + tau_se )
                    + 2. * J_ei**2 * x[7] * tau_si * taum / ( (1. + rei ) * taum + tau_si )
                    + sigma_ext_e**2 )
    G[4] = - interpolate(tau_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma)
    G[1] = - interpolate(r_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma) * 1e3
    rii = rxy(cii, Ki, tau_si, z[1], J_ii)
    rie = rxy(cie, Ke, tau_se, z[0], J_ie)
    G[3] = - np.sqrt( 2. * J_ii**2 * x[9] * tau_si * taum / ( (1. + rii ) * taum + tau_si )
                    + 2. * J_ie**2 * x[8] * tau_se * taum / ( (1. + rie ) * taum + tau_se )
                    + sigma_ext_i**2 )
    G[5] = - interpolate(tau_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma)
    return G
```

```python
def interpolate(table_, mu_, sigma_, mu_range_, d_mu_, sigma_range_, d_sigma_):
    xid1, yid1, dxid, dyid = timeIntegration.fast_interp2_opt(sigma_range_, d_sigma_, sigma_, mu_range_, d_mu_, mu_)
    xid1, yid1 = int(xid1), int(yid1)
    return timeIntegration.interpolate_values(table_, xid1, yid1, dxid, dyid)
    
def rate_e(x, z):
    return interpolate(r_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma) * 1e3

def rate_i(x, z):
    return interpolate(r_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma) * 1e3

def tau_e(x, z):
    return interpolate(tau_table, x[0], z[2], mu_range, d_mu, sigma_range, d_sigma)

def tau_i(x, z):
    return interpolate(tau_table, x[1], z[3], mu_range, d_mu, sigma_range, d_sigma)

def rxy(cxy, Ky, tau_sy, ry, J_xy):
    return cxy * Ky * tau_sy * ry * 1e-3 / np.abs(J_xy)

def pxy(cxy, Ky, tau_sy, ry, J_xy):
    return cxy**2 * Ky * tau_sy**2 * ry * 1e-3 / J_xy**2
```

```python
### SIMULATION

def fp(x_timeseries, z_timeseries):
    x = np.zeros((dim_x))
    z = np.zeros((dim_z))
    x[:] = x_timeseries[:,-1]
    z[:] = z_timeseries[:,-1]
    
    return x, z

def run_sim(t, dt, final='low', plot=False):
    
    time = np.arange(0,t,dt)

    x_timeseries = np.zeros(( dim_x, int(t/dt) ))
    z_timeseries = np.zeros(( dim_z, int(t/dt) ))

    if final == 'low':
        setinitlow(x_timeseries, z_timeseries)
    elif final == 'high':
        setinithigh(x_timeseries, z_timeseries)
    elif final=='random':
        setinitrandom(x_timeseries, z_timeseries)
        
    for t_ in range(1, int(t/dt)):
        F_timeseries = F(x_timeseries[:,t_-1], z_timeseries[:,t_-1])
        for j in range(dim_x):
            x_timeseries[j,t_] = x_timeseries[j,t_-1] + dt * (F_timeseries[j])

        G_timeseries = G(x_timeseries[:,t_-1], z_timeseries[:,t_-1])
        for j in range(dim_z):
            z_timeseries[j,t_] = - G_timeseries[j]

    if plot:
        fig, ax = plt.subplots(5, 2, figsize=(15,10))

        ax[0,0].plot(time, x_timeseries[0,:], label='x0=mu_e', color='red')
        ax[0,0].plot(time, x_timeseries[1,:], label='x1=mu_i', color='blue')
        ax[1,0].plot(time, x_timeseries[2,:], label='x2=see', color='red')
        ax[1,0].plot(time, x_timeseries[3,:], label='x3=sei', color='blue')
        ax[2,0].plot(time, x_timeseries[4,:], label='x4=sie', color='red')
        ax[2,0].plot(time, x_timeseries[5,:], label='x5=sii', color='blue')
        ax[3,0].plot(time, x_timeseries[6,:], label='x6=sigma_ee', color='red')
        ax[3,0].plot(time, x_timeseries[7,:], label='x7=sigma_ei', color='blue')
        ax[4,0].plot(time, x_timeseries[8,:], label='x8=sigma_ie', color='red')
        ax[4,0].plot(time, x_timeseries[9,:], label='x9=sigma_ii', color='blue')

        ax[0,1].plot(time, z_timeseries[0,:], label='z0=r_e', color='red')
        ax[1,1].plot(time, z_timeseries[1,:], label='z1=r_i', color='blue')
        ax[2,1].plot(time, z_timeseries[2,:], label='z2=sigma_e', color='red')
        ax[3,1].plot(time, z_timeseries[3,:], label='z3=sigma_i', color='blue')
        ax[4,1].plot(time, z_timeseries[4,:], label='z4=tau_e', color='red')
        ax[4,1].plot(time, z_timeseries[5,:], label='z5=tau_i', color='blue')

        for j1 in range(5):
                ax[j1,0].legend()
                if j1 < dim_z:
                    ax[j1,1].legend()

        plt.show()
    
    x, z = fp(x_timeseries, z_timeseries)
    
    return x, z
    
x, z = run_sim(10., 0.1, final='low', plot=True)

print('fixed point:')
print(x)
print(z)
```

```python
def F_dx(x, z, dx):
    F_dx = np.zeros(( x.shape[0], x.shape[0] ))
    for i in range(x.shape[0]):
        for j in range(F_dx.shape[1]):
            x_dx = x.copy()
            x_dx[j] = x[j] + dx
            F_1 = F(x_dx, z)
            F_0 = F(x, z)
            F_dx[i,j] = ( F_1[i] - F_0[i] ) / (dx)
    return F_dx

def F_dz(x, z, dz):
    F_dz = np.zeros(( x.shape[0], z.shape[0] ))
    for i in range(F_dz.shape[0]):
        for j in range(F_dz.shape[1]):
            z_dz = z.copy()
            z_dz[j] = z[j] + dz
            F_1 = F(x, z_dz)
            F_0 = F(x, z)
            F_dz[i,j] = ( F_1[i] - F_0[i] ) / (dz)
    return F_dz

def G_dx(x, z, dx):
    G_dx = np.zeros(( z.shape[0], x.shape[0] ))
    for i in range(G_dx.shape[0]):
        for j in range(G_dx.shape[1]):
            x_dx = x.copy()
            x_dx[j] = x[j] + dx
            G_1 = G(x_dx, z)
            G_0 = G(x, z)
            G_dx[i,j] = ( G_1[i] - G_0[i] ) / (dx)
    return G_dx

def G_dz(x, z, dz):
    G_dz = np.zeros(( z.shape[0], z.shape[0] ))
    for i in range(G_dz.shape[0]):
        for j in range(G_dz.shape[1]):
            z_dz = z.copy()
            z_dz[j] = z[j] + dz
            G_1 = G(x, z_dz)
            G_0 = G(x, z)
            G_dz[i,j] = ( G_1[i] - G_0[i] ) / (dz)
        G_dz[i,i] += 1.
    return G_dz
```

```python
def H(x, z, dx, dz):
    H0 = G_dz(x, z, dx)
    H1 = np.linalg.inv(H0)
    H2 = G_dx(x, z, dx)
    H3 = F(x, z)
    H12 = np.tensordot(H1, H2, axes=1)
    H123 = np.tensordot(H12, H3, axes=1)
    return - H123

def H_dx(x, z, dx, dz):
    H_dx = np.zeros((dim_z, dim_x))
    for i in range(H_dx.shape[0]):
        for j in range(H_dx.shape[1]):
            x_dx = x.copy()
            x_dx[j] = x[j] + dx
            H_1 = H(x_dx, z, dx, dz)
            H_0 = H(x, z, dx, dz)
            H_dx[i,j] = ( H_1[i] - H_0[i] ) / (dx)
    return H_dx

def H_dz(x, z, dx, dz):
    H_dz = np.zeros((dim_z, dim_z))
    for i in range(H_dz.shape[0]):
        for j in range(H_dz.shape[1]):
            z_dz = z.copy()
            z_dz[j] = z[j] + dz
            H_1 = H(x, z_dz, dx, dz)
            H_0 = H(x, z, dx, dz)
            H_dz[i,j] = ( H_1[i] - H_0[i] ) / (dz)
    return H_dz
```

```python
dx = 0.0001
dt = 0.1
t = 100.
x_, z_ = run_sim(t, dt, final='low')
#x_, z_ = ufp()

def jacobian(x, z, dx):

    F1 = F_dx(x,z, dx)
    F2 = F_dz(x, z, dx)
    #H0 = H(x, z, dx, dx)
    H1 = H_dx(x, z, dx, dx)
    H2 = H_dz(x, z, dx, dx)

    jac = np.zeros((dim_x+dim_z, dim_x+dim_z))
    jac[:dim_x,:dim_x] = F1
    jac[dim_x:, :dim_x] = H1
    jac[:dim_x, dim_x:] = F2
    jac[dim_x:, dim_x:] = H2

    #for j1 in []:
    #    jac = np.delete(jac, j1, 0)
    #    jac = np.delete(jac, j1, 1)
        
    return jac

jac = jacobian(x_, z_, dx)

#print('round 8 ----------')
#print(np.around(np.linalg.eigvals(jac), decimals=8))
#print('round 6 ----------')
#print(np.around(np.linalg.eigvals(jac), decimals=6))
#print('round 4 ----------')
print(np.around(np.linalg.eigvals(jac), decimals=4))
```

```python
# low, high state, should be stable

dx = 1e-5
t = 100.
dt = 0.1

max_pos_l = 0.
max_pos_ind_l = -1

max_pos_h = 0.
max_pos_ind_h = -1

zero_real = []

latex_string_low_ev = ''
latex_string_high_ev = ''

limit = 1e-10

for i in range_:
    print('-----------------------------')
    print(i, ', external input: ', ext_exc[i], ext_inh[i])
    mu_ext_e = ext_exc[i] * 5.
    mu_ext_i = ext_inh[i] * 5.
    
    for state in ['low', 'high']:
    
        x_, z_ = run_sim(t, dt, final=state, plot=False)
        
        if state == 'low' and z_[0] > 10.:
            print('not in low state')
        elif state == 'high' and z_[0] < 10.:
            print('not in high state')
        
        jac = jacobian(x_, z_, dx)
        ev = np.linalg.eigvals(jac)
        
        if state == 'low':
            jac_min = jacobian(x_, z_, 10. * dx)
        else:
            jac_min = jacobian(x_, z_, -dx)
        ev_min = np.linalg.eigvals(jac_min)
        ev_diff = ev - ev_min
        
        imag_ind = []
        ev_sorted = ev.copy()
        index_list = []
        
        for j in range(len(ev_sorted)):
            if np.abs(ev_sorted[j].imag) > limit:
                ev_sorted[j] = 100.
                imag_ind.append(j)

        ev_sorted = np.array( sorted(ev_sorted) )
        for j in range(len(ev_sorted)):
            if ev_sorted[j] < 100.:
                index_list.append(np.where(ev == ev_sorted[j])[0][0])

        index_comp = len(ev_sorted) - len(imag_ind)

        for j in imag_ind:
            for k in imag_ind:
                #print(j, k) 
                if j >= k:
                    continue
                if (np.abs(ev[j].imag + ev[k].imag) < 1e-4 * limit) and (np.abs(ev[j].real - ev[k].real) < 1e-4 * limit):
                    #print(j, k)
                    ev_sorted[index_comp] = ev[j]
                    ev_sorted[index_comp+1] = ev[k]
                    index_comp += 2
                    index_list.append(j)
                    index_list.append(k)
                
        ev = ev_sorted.copy()
        
        #print('index list = ', index_list)
        
        if state == 'low':
            latex_string_low_ev = (latex_string_low_ev + '{:.3f}'.format(mu_ext_e) +
                        ' & ' + '{:.3f}'.format(mu_ext_i) + ' & ')
            j = 0
            for e_ in ev:
                opacity = 0.
                if np.abs(e_) > limit:
                    error = min(100. * 2. * np.abs(ev_diff[index_list[j]]) / np.abs(e_), 100)
                    opacity = int(100-error/2.)
                    #print('error, opacity = ', error, opacity)
                j += 1
                if np.abs(e_.real) >= 1.:
                    latex_string_low_ev = ( latex_string_low_ev
                                           + '{:.2f}'.format(e_.real) )
                elif e_.real >= limit:
                    if opacity <= 50.:
                        latex_string_low_ev = ( latex_string_low_ev + '\\textbf{'
                                           + '{:.2e}'.format(e_.real) + '}')
                    else:
                        latex_string_low_ev = ( latex_string_low_ev + '\cellcolor{red!' + str(opacity) + '}'
                                           + '{:.2e}'.format(e_.real) )
                        print("Value, error: ", e_, ev_diff[index_list[j-1]] )
                elif e_.real <= -limit:
                    if opacity <= 50.:
                        latex_string_low_ev = ( latex_string_low_ev + '\\textbf{'
                                           + '{:.2e}'.format(e_.real) + '}')
                    else:
                        latex_string_low_ev = ( latex_string_low_ev + '{:.2e}'.format(e_.real) )
                else:
                    latex_string_low_ev = latex_string_low_ev + '0.00'
                    
                if e_.imag >= 1.:
                    latex_string_low_ev = ( latex_string_low_ev
                                           + ' + {:.2f}'.format(e_.imag) + 'i & ' )
                elif limit <= e_.imag < 1.:
                    latex_string_low_ev = ( latex_string_low_ev
                                           + ' + {:.2e}'.format(e_.imag) + 'i & ' )
                elif -limit >= e_.imag > -1.:
                    latex_string_low_ev = ( latex_string_low_ev 
                                           + ' - {:.2e}'.format(np.abs(e_.imag)) + 'i & ' )
                elif e_.imag <= -1.:
                    latex_string_low_ev = ( latex_string_low_ev
                                           + ' - {:.2f}'.format(np.abs(e_.imag)) + 'i & ' )
                else:
                    latex_string_low_ev = ( latex_string_low_ev + ' & ' )

            latex_string_low_ev = latex_string_low_ev[:-2] + ' \\\ \n'
            
        else:
            latex_string_high_ev = (latex_string_high_ev + '{:.3f}'.format(mu_ext_e) +
                        ' & ' + '{:.3f}'.format(mu_ext_i) + ' & ')
            j = 0
            for e_ in ev:
                opacity = 0.
                if np.abs(e_) > limit:
                    error = min(100. * 2. * np.abs(ev_diff[index_list[j]]) / np.abs(e_), 100)
                    opacity = int(100-error/2.)
                j += 1
                if np.abs(e_.real) >= 1.:
                    latex_string_high_ev = ( latex_string_high_ev
                                           + '{:.2f}'.format(e_.real) )
                elif e_.real >= limit:
                    if opacity <= 50.:
                        latex_string_high_ev = ( latex_string_high_ev + '\\textbf{'
                                           + '{:.2e}'.format(e_.real) + '}')
                    else:
                        latex_string_high_ev = ( latex_string_high_ev + '\cellcolor{red!' + str(50) + '}'
                                           + '{:.2e}'.format(e_.real) )
                        print("Value, error: ", e_, ev_diff[index_list[j-1]] )
                elif e_.real <= -limit:
                    if opacity <= 50.:
                        latex_string_high_ev = ( latex_string_high_ev + '\\textbf{'
                                           + '{:.2e}'.format(e_.real) + '}')
                    else:
                        latex_string_high_ev = ( latex_string_high_ev + '{:.2e}'.format(e_.real) )
                else:
                    latex_string_high_ev = latex_string_high_ev + '0.00'
                    
                if e_.imag >= 1.:
                    latex_string_high_ev = ( latex_string_high_ev
                                           + ' + {:.2f}'.format(e_.imag) + 'i & ' )
                elif limit <= e_.imag < 1.:
                    latex_string_high_ev = ( latex_string_high_ev
                                           + ' + {:.2e}'.format(e_.imag) + 'i & ' )
                elif -limit >= e_.imag > -1.:
                    latex_string_high_ev = ( latex_string_high_ev 
                                           + ' - {:.2e}'.format(np.abs(e_.imag)) + 'i & ' )
                elif e_.imag <= -1.:
                    latex_string_high_ev = ( latex_string_high_ev
                                           + ' - {:.2f}'.format(np.abs(e_.imag)) + 'i & ' )
                else:
                    latex_string_high_ev = ( latex_string_high_ev + ' & ' )

            latex_string_high_ev = latex_string_high_ev[:-2] + ' \\\ \n'

        #print(state, ', positive real parts:')
        for re in ev_sorted:
            if np.abs(re) < limit:
                #print('small real part:', re)
                if i not in zero_real:
                    zero_real.append(i)
        
        max_positive = ev_sorted[-1].real
        
        if state == 'low':
            if max_positive > max_pos_l and max_positive > 0.:
                max_pos_l = max_positive
                max_pos_ind_l = i
            elif max_positive < 0.:
                print("no positive EV, ", i)
        else:
            if max_positive > max_pos_h and max_positive > 0.:
                max_pos_h = max_positive
                max_pos_ind_h = i
            elif max_positive < 0.:
                print("no positive EV, ", i)
        
#latex_string_low_ev = latex_string_low_ev.replace('e', '')
            
print('maximum positive value:')
print(max_pos_ind_l, max_pos_l)
print(max_pos_ind_h, max_pos_h)

print('zero real part: ', zero_real)
```

```python
def setxz(state):
    #print(state)
    x = np.zeros((dim_x))
    z = np.zeros((dim_z))
    
    x[0] = state[2]
    x[1] = state[3]
    x[2] = state[5]
    x[3] = state[6]
    x[4] = state[7]
    x[5] = state[8]
    x[6] = state[9]
    x[7] = state[10]
    x[8] = state[11]
    x[9] = state[12]
    
    z[0] = state[0]
    z[1] = state[1]
    z[2] = state[15]
    z[3] = state[16]
    z[4] = state[18]
    z[5] = state[19]
    
    return x, z

def readcase(case_, range_):
    
    global ufp_full
    
    readpath = 'data_final' + os.sep
    
    control_data = data.read_control(readpath, case_)
    opt_state = control_data[3]

    ufp_full = np.zeros(( 147, 20 ))
    
    return_range = []
        
    for i in range_:
                
        if type(opt_state[i]) == type(None):
            continue
            
        return_range.append(i)
            
        if case_ in ['1', '2']:
            ufp, ufp_fulldata = data.get_ufp(opt_state[i], 0)
        else:
            ufp, ufp_fulldata = data.get_ufp(opt_state[i], 2)
        
        if ufp[0] > 1e-1  and ufp[1] > 1e-1:
            ufp_full[i,:] = ufp_fulldata
            
    #return return_range
```

```python
# unstable fixed point, should be unstable

dx = 1e-5

min_pos = []
max_pos = []

zero_real = []
nonzero_real = []

latex_string_plateau = ''

counter = 0
limit = 1e-10

for i in range_:
    print('-----------------------------')
    print(i, ', external input: ', ext_exc[i], ext_inh[i])
    #print('unstable state rates: ', ufp_full[i][:2])
    
    mu_ext_e = ext_exc[i] * 5.
    mu_ext_i = ext_inh[i] * 5.
    
    if ufp_full[i][0] == 0.:
        continue
        
    counter += 1
        
    x, z = setxz(ufp_full[i])
        
    jac = jacobian(x, z, dx)
    ev = np.linalg.eigvals(jac)
    ev_sorted_real = np.array( sorted(ev, key=lambda x: x.real) )
    
    jac_min = jacobian(x, z, - dx)
    ev_min = np.linalg.eigvals(jac_min)
    ev_diff = ev - ev_min
    
    jac_diff = jac - jac_min
    
    print('jac ev diff max = ', np.amax(np.abs(ev_diff)) )
        
    imag_ind = []
    ev_sorted = ev.copy()
    index_list = []
        
    for j in range(len(ev_sorted)):
        if np.abs(ev_sorted[j].imag) > limit:
            ev_sorted[j] = 100.
            imag_ind.append(j)

    ev_sorted = np.array( sorted(ev_sorted) )
    for j in range(len(ev_sorted)):
        if ev_sorted[j] < 100.:
            index_list.append(np.where(ev == ev_sorted[j])[0][0])

    index_comp = len(ev_sorted) - len(imag_ind)

    for j in imag_ind:
        for k in imag_ind:
            #print(j, k) 
            if j >= k:
                continue
            if (np.abs(ev[j].imag + ev[k].imag) < 1e-4 * limit) and (np.abs(ev[j].real - ev[k].real) < 1e-4 * limit):
                #print(j, k)
                ev_sorted[index_comp] = ev[j]
                ev_sorted[index_comp+1] = ev[k]
                index_comp += 2
                index_list.append(j)
                index_list.append(k)
                
    ev = ev_sorted.copy()
    

    for re in ev_sorted:
        if np.abs(re) < limit:
            print('small real part:', re)
            if i not in zero_real:
                zero_real.append(i)
            if i in nonzero_real:
                nonzero_real.pop()
        else:
            if i not in nonzero_real and i not in zero_real:
                nonzero_real.append(i)
                
    latex_string_plateau = (latex_string_plateau + '{:.3f}'.format(mu_ext_e) +
                        ' & ' + '{:.3f}'.format(mu_ext_i) + ' & ')
    
    j = 0
    
    for e_ in ev:
        opacity = 0.
        if np.abs(e_) > limit:
            error = min(100. * 2. * np.abs(ev_diff[index_list[j]]) / np.abs(e_), 100)
            opacity = int(100-error/2.)
            if error >= 100:
                print('value, error : ', e_.real, error)
        j += 1
        if np.abs(e_.real) >= 1.:
            if opacity <= 50:
                latex_string_plateau = ( latex_string_plateau + '\\textbf{' 
                                        + '{:.2f}'.format(e_.real) + '}' )
            else:
                if e_.real > 0.:
                    latex_string_plateau = ( latex_string_plateau + '\cellcolor{green!' + str(50) + '}'
                                            + '{:.2f}'.format(e_.real) )
                    print('green')
                else:
                    latex_string_plateau = latex_string_plateau + '{:.2f}'.format(e_.real)
        elif np.abs(e_.real) >= 1e-10:
            if opacity <= 50:
                latex_string_plateau = ( latex_string_plateau + '\\textbf{' 
                                        + '{:.2e}'.format(e_.real) + '}' )
            else:
                if e_.real > 0.:
                    latex_string_plateau = ( latex_string_plateau + '\cellcolor{green!' + str(50) + '}'
                                            + '{:.2e}'.format(e_.real) )
                    print('green')
                else:
                    latex_string_plateau = latex_string_plateau + '{:.2e}'.format(e_.real)
        else:
            latex_string_plateau = latex_string_plateau + '0.00'

        if e_.imag >= 1.:
            latex_string_plateau = latex_string_plateau + ' + {:.2f}'.format(e_.imag) + 'i & '
        elif 1e-10 <= e_.imag < 1.:
            latex_string_plateau = latex_string_plateau + ' + {:.2e}'.format(e_.imag) + 'i & '
        elif -1e-10 >= e_.imag > -1.:
            latex_string_plateau = latex_string_plateau + ' - {:.2e}'.format(np.abs(e_.imag)) + 'i & '
        elif e_.imag <= -1.:
            latex_string_plateau = latex_string_plateau + ' - {:.2f}'.format(np.abs(e_.imag)) + 'i & '
        else:
            latex_string_plateau = latex_string_plateau + ' & '

    latex_string_plateau = latex_string_plateau[:-2] + ' \\\ \n'

    max_pos.append(ev_sorted_real[-1].real)
    for e_ in ev_sorted_real:
        if e_.real > 1e-10:
            min_pos.append(e_.real)
            break

print('total number of points = ', counter)
print('maximum positive value:')
print( max(max_pos), max_pos.index(max(max_pos)) )
print('minimum positive value:')
print( min(min_pos), min_pos.index(min(min_pos)) )
print('zero real part: ', zero_real)
print('non zero real part: ', nonzero_real)
```
