### Bistability

```python
aln = ALNModel()
data.set_parameters(aln)
aln.params.tau_se = 600.
aln.params.tau_si = 150.

state_vars = aln.state_vars
init_vars = aln.init_vars

##############################################################
def setinit(init_vars_):
    for iv in range(len(init_vars)):
        for sv in range(len(state_vars)):
            if state_vars[sv] in init_vars[iv]:
                #print("set init vars ", )
                if aln.params[init_vars[iv]].ndim == 2:
                    aln.params[init_vars[iv]][0,:] = init_vars_[sv]
                else:
                    aln.params[init_vars[iv]][0] = init_vars_[sv]
```

```python
step = 0.1
range_e = np.arange(0.45, 0.451, step)
range_i = np.arange(0.475, 0.4751, step)

aln.params['duration'] = 6000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    bistable = False
    for e in range_e:
        #print('---------', e)
        bistable = False
        
        aln.params.ext_exc_current = e * 5.
        aln.params.ext_inh_current = i * 5.
        
        aln.run(control=control0)
            
        print(e, i)
        plt.plot(aln.t, aln.rates_exc[0,:], color='red')
        plt.plot(aln.t, aln.rates_inh[0,:], color='blue')
        plt.show()
```

```python
lower_bound_bi_e = []
lower_bound_bi_i = []
upper_bound_bi_e = []
upper_bound_bi_i = []
bi_e = []
bi_i = []
```

```python
step = 0.005
range_e = np.arange(0.2, 0.501, step)
range_i = np.arange(0.1, 0.901, step)

aln.params['duration'] = 6000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    bistable = False
    for e in range_e:
        #print('---------', e)
        bistable = False
        
        aln.params.mue_ext_mean = e * 5.
        aln.params.mui_ext_mean = i * 5.
        
        aln.run(control=control0)
        #plt.plot(aln.t, aln.rates_exc[0,:], color='red')
        #plt.show()
        
        mean1_1 = np.mean(aln.rates_exc[0,27500:28000])
        mean1_2 = np.mean(aln.rates_exc[0,28500:29000])
        mean1_3 = np.mean(aln.rates_exc[0,29500:30000])
        
        if np.abs(mean1_1 - mean1_2) < 1e-4 and np.abs(mean1_1 - mean1_3) < 1e-4:
            stable1 = True
        
        mean2_1 = np.mean(aln.rates_exc[0,57500:58000])
        mean2_2 = np.mean(aln.rates_exc[0,58500:59000])
        mean2_3 = np.mean(aln.rates_exc[0,59500:60000])
        
        if np.abs(mean2_1 - mean2_2) < 1e-4 and np.abs(mean2_1 - mean2_3) < 1e-4:
            stable2 = True
            
        if stable1 and stable2 and np.abs(mean1_1 - mean2_1) > 1.:
            bistable = True
            
        #print(bistable, lower_bound, upper_bound)
            
        if bistable:
            
            upper_bound = True
            bi_e.append(e)
            bi_i.append(i)
            
            #print(bi_e[-1], bi_i[-1])
            #plotFunc.plot_traces(aln, control0)
            #plt.show()
            
            #if np.abs(bi_e[-1] - 0.4) < 1e-4 and np.abs(bi_i[-1] - 0.325) < 1e-4:
            #        bi_e.pop()
            #        bi_i.pop()
            
            if lower_bound == False:
                lower_bound_bi_e.append(e)
                lower_bound_bi_i.append(i)
                lower_bound = True
                # remove osc/ up bistable state
                #if np.abs(lower_bound_bi_e[-1] - 0.4) < 1e-4 and np.abs(lower_bound_bi_i[-1] - 0.325) < 1e-4:
                #    lower_bound_bi_e.pop()
                #    lower_bound_bi_i.pop()
                
        if not bistable:
            if upper_bound:
                upper_bound = False
                upper_bound_bi_e.append(e-step)
                upper_bound_bi_i.append(i)
                # remove osc/ up bistable state
                #if np.abs(upper_bound_bi_e[-1] - 0.4) < 1e-4 and np.abs(upper_bound_bi_i[-1] - 0.325) < 1e-4:
                #    upper_bound_bi_e.pop()
                #    upper_bound_bi_i.pop()
                
            
plt.scatter(bi_e, bi_i, s=60, marker='x', c="grey")
plt.scatter(lower_bound_bi_e, lower_bound_bi_i, s=12, c="red")
plt.scatter(upper_bound_bi_e, upper_bound_bi_i, s=12, c="blue")
plt.show()

val = []
for i,j in zip(bi_e, bi_i):
    if [i,j] in val:
        print("already there", i,j)
    val.append([i,j])
```

```python
lower_bound_LC_e = []
lower_bound_LC_i = []
upper_bound_LC_e = []
upper_bound_LC_i = []
LC_e = []
LC_i = []
```

```python
step = 0.01
range_e = np.arange(0.0, 0.231, step)
range_i = np.arange(0.0, 0.121, step)

aln.params['duration'] = 24000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    for e in range_e:
        print('---------', e, i)
        stable1 = False
        stable2 = False
        stable3 = False
        
        aln.params.mue_ext_mean = e * 5.
        aln.params.mui_ext_mean = i * 5.
        
        aln.run(control=control0)
        
        mean1_1 = np.mean(aln.rates_exc[0,108050:112000])
        mean1_2 = np.mean(aln.rates_exc[0,112050:116000])
        mean1_3 = np.mean(aln.rates_exc[0,116050:120000])
        #print(mean1_1, mean1_2, mean1_3)
        if np.abs(mean1_1 - mean1_2) < 1e-4 and np.abs(mean1_1 - mean1_3) < 1e-4:
            stable1 = True
        
        mean2_1 = np.mean(aln.rates_exc[0,228050:1232000])
        mean2_2 = np.mean(aln.rates_exc[0,232050:236000])
        mean2_3 = np.mean(aln.rates_exc[0,236050:240000])
                
        if np.abs(mean2_1 - mean2_2) < 1e-4 and np.abs(mean2_1 - mean2_3) < 1e-4:
            stable2 = True
            
        if stable1 and stable2 and np.abs(mean1_1 - mean2_1) < 1e-4:
            stable3 = True
            
        plt.plot(aln.t, aln.rates_exc[0,:], color='red')
        plt.plot(aln.t, aln.rates_inh[0,:], color='blue')
        plt.ylim(0.3, 0.4)
        plt.show()
            
        if not stable3:
            #plotFunc.plot_traces(aln, control0)
            upper_bound = True
            LC_e.append(e)
            LC_i.append(i)
            print("oscillations, ", e, i)
            #plt.plot(aln.t, aln.rates_exc[0,:], color='red')
            #plt.plot(aln.t, aln.rates_inh[0,:], color='blue')
            #plt.ylim(0., 1.)
            #plt.show()
            
            #break
            
            if lower_bound == False:
                lower_bound_LC_e.append(e)
                lower_bound_LC_i.append(i)
                lower_bound = True
                
        if stable3:
            if upper_bound:
                upper_bound = False
                upper_bound_LC_e.append(e-step)
                upper_bound_LC_i.append(i)
                
                print("upper bound, ", e, i)
                plt.plot(aln.t, aln.rates_exc[0,:], color='red')
                plt.plot(aln.t, aln.rates_inh[0,:], color='blue')
                plt.ylim(0., 10.)
                plt.show()
                
        #print(bi_e, lower_bound_bi_e, upper_bound_bi_e)
        
    if len(LC_e) >= 1:
        break
                
            
plt.scatter(LC_e, LC_i, s=60, marker='x', c="grey")
plt.scatter(lower_bound_LC_e, lower_bound_LC_i, s=12, c="red")
plt.scatter(upper_bound_LC_e, upper_bound_LC_i, s=12, c="blue")

# 0.325, 0.275 is osc
# 0.35, 0.275 is osc
# 0.35, 0.3 is down state
# 0.375, 0.3 is osc
# 0.35, 0.325 is down state
# 0.375, 0.325 is down state
# 0.4, 0.325 is bistable LC, up
# 0.425, 0.325 is up
```
