```python
lower_bound_bi_e = []
lower_bound_bi_i = []
upper_bound_bi_e = []
upper_bound_bi_i = []
bi_e = []
bi_i = []
```

```python
step = 0.001
range_e = np.arange(0.3, 0.81, step)
range_i = np.arange(0.33, 0.701, step)

aln.params['duration'] = 6000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    bistable = False
    for e in range_e:
        #print('---------', e)
        bistable = False
        
        aln.params.mue_ext_mean = e * 5.
        aln.params.mui_ext_mean = i * 5.
        
        aln.run(control=control0)
        
        mean1_1 = np.mean(aln.rates_exc[0,27500:28000])
        mean1_2 = np.mean(aln.rates_exc[0,28500:29000])
        mean1_3 = np.mean(aln.rates_exc[0,29500:30000])
        
        if np.abs(mean1_1 - mean1_2) < 1e-4 and np.abs(mean1_1 - mean1_3) < 1e-4:
            stable1 = True
        
        mean2_1 = np.mean(aln.rates_exc[0,57500:58000])
        mean2_2 = np.mean(aln.rates_exc[0,58500:59000])
        mean2_3 = np.mean(aln.rates_exc[0,59500:60000])
        
        if np.abs(mean2_1 - mean2_2) < 1e-4 and np.abs(mean2_1 - mean2_3) < 1e-4:
            stable2 = True
            
        if stable1 and stable2 and np.abs(mean1_1 - mean2_1) > 1.:
            bistable = True
            
        #print(bistable, lower_bound, upper_bound)
            
        if bistable:
            
            upper_bound = True
            bi_e.append(e)
            bi_i.append(i)
            
            #print(bi_e[-1], bi_i[-1])
            #plotFunc.plot_traces(aln, control0)
            #plt.show()
            
            if np.abs(bi_e[-1] - 0.4) < 1e-4 and np.abs(bi_i[-1] - 0.325) < 1e-4:
                    bi_e.pop()
                    bi_i.pop()
            
            if lower_bound == False:
                lower_bound_bi_e.append(e)
                lower_bound_bi_i.append(i)
                lower_bound = True
                # remove osc/ up bistable state
                #if np.abs(lower_bound_bi_e[-1] - 0.4) < 1e-4 and np.abs(lower_bound_bi_i[-1] - 0.325) < 1e-4:
                #    lower_bound_bi_e.pop()
                #    lower_bound_bi_i.pop()
                
        if not bistable:
            if upper_bound:
                upper_bound = False
                upper_bound_bi_e.append(e-step)
                upper_bound_bi_i.append(i)
                # remove osc/ up bistable state
                #if np.abs(upper_bound_bi_e[-1] - 0.4) < 1e-4 and np.abs(upper_bound_bi_i[-1] - 0.325) < 1e-4:
                #    upper_bound_bi_e.pop()
                #    upper_bound_bi_i.pop()
                
            
plt.scatter(bi_e, bi_i, s=30, marker='x', c="grey")
plt.scatter(lower_bound_bi_e, lower_bound_bi_i, s=12, c="red")
plt.scatter(upper_bound_bi_e, upper_bound_bi_i, s=12, c="blue")

val = []
for i,j in zip(bi_e, bi_i):
    if [i,j] in val:
        print("already there", i,j)
    val.append([i,j])
```

```python
boundary_bi_e = []
boundary_bi_i = []

for index in range(len(upper_bound_bi_e)-1, -1, -1):
    boundary_bi_e.append(upper_bound_bi_e[index])
    boundary_bi_i.append(upper_bound_bi_i[index])
for index in range(len(lower_bound_bi_e)):
    boundary_bi_e.append(lower_bound_bi_e[index])
    boundary_bi_i.append(lower_bound_bi_i[index])
    
plt.scatter(boundary_bi_e, boundary_bi_i, s=12, c="blue")
```

```python
range_e = np.arange(0.0, 0.51, step)
range_i = np.arange(0.0, 0.321, step)

aln.params['duration'] = 6000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    for e in range_e:
        #print('---------', e)
        stable1 = False
        stable2 = False
        stable3 = False
        
        aln.params.mue_ext_mean = e * 5.
        aln.params.mui_ext_mean = i * 5.
        
        aln.run(control=control0)
        
        mean1_1 = np.mean(aln.rates_exc[0,27050:28000])
        mean1_2 = np.mean(aln.rates_exc[0,28050:29000])
        mean1_3 = np.mean(aln.rates_exc[0,29050:30000])
        #print(mean1_1, mean1_2, mean1_3)
        if np.abs(mean1_1 - mean1_2) < 1e-4 and np.abs(mean1_1 - mean1_3) < 1e-4:
            stable1 = True
        
        mean2_1 = np.mean(aln.rates_exc[0,57050:58000])
        mean2_2 = np.mean(aln.rates_exc[0,58050:59000])
        mean2_3 = np.mean(aln.rates_exc[0,59050:60000])
                
        if np.abs(mean2_1 - mean2_2) < 1e-4 and np.abs(mean2_1 - mean2_3) < 1e-4:
            stable2 = True
            
        if stable1 and stable2 and np.abs(mean1_1 - mean2_1) < 1e-4:
            stable3 = True
            
        #print(stable1, stable2, stable3)
            
        if not stable3:
            #plotFunc.plot_traces(aln, control0)
            upper_bound = True
            LC_e.append(e)
            LC_i.append(i)
            
            if lower_bound == False:
                lower_bound_LC_e.append(e)
                lower_bound_LC_i.append(i)
                lower_bound = True
                
        if stable3:
            if upper_bound:
                upper_bound = False
                upper_bound_LC_e.append(e-step)
                upper_bound_LC_i.append(i)
                
        #print(bi_e, lower_bound_bi_e, upper_bound_bi_e)
                
            
plt.scatter(LC_e, LC_i, s=60, marker='x', c="grey")
plt.scatter(lower_bound_LC_e, lower_bound_LC_i, s=12, c="red")
plt.scatter(upper_bound_LC_e, upper_bound_LC_i, s=12, c="blue")

# 0.325, 0.275 is osc
# 0.35, 0.275 is osc
# 0.35, 0.3 is down state
# 0.375, 0.3 is osc
# 0.35, 0.325 is down state
# 0.375, 0.325 is down state
# 0.4, 0.325 is bistable LC, up
# 0.425, 0.325 is up
```
