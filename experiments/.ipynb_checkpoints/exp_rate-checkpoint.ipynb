```python
from neurolib_dev_optControl.neurolib.models.aln import ALNModel
import neurolib_dev_optControl.neurolib.dashboard.data as data

import matplotlib.pyplot as plt
import numpy as np
    
%load_ext autoreload
%autoreload 2 
```

```python
aln = ALNModel()
data.set_parameters(aln)

aln.params.duration = 2000.
control0 = aln.getZeroControl()

aln.params.mui_ext_mean = 0.5 * 5.
aln.params.mue_ext_mean = 0.54 * 5.

control0[0,0,0:1000] = 3.
control0[0,0,5000:7000] = -3.
#control0[0,3,5000:15000] = 1.

aln.run(control=control0)

plt.plot(aln.t, aln.rates_exc[0,:])
plt.plot(aln.t, aln.rates_inh[0,:])
plt.show()
```

```python
aln = ALNModel()
data.set_parameters(aln)

aln.params.dt = 0.01
aln.params.duration = 2000.
control0 = aln.getZeroControl()

aln.params.mui_ext_mean = 0.5 * 5.
aln.params.mue_ext_mean = 0.45 * 5.

control0[0,0,0:10000] = 3.
#control0[0,0,5000:7000] = -3.
control0[0,3,50000:150000] = 1e5

aln.run(control=control0)

plt.plot(aln.t, aln.rates_exc[0,:])
plt.plot(aln.t, aln.rates_inh[0,:])
plt.show()
```
